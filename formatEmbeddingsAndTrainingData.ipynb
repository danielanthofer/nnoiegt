# Format Embeddings

### Regular Words

The embeddings produced by GloVe are too many, therefore, we only select embeddings for words that appear at least 4 times in the training data set. Any other word will be replaced by a special placeholder word. For now, we will allow 5 replacements, any other word will be replaced with "unknown"

```python
# Modified Input file to count only word and lemma

# modified conll-X table row elements
i_ID = 0
i_FORM = 1
i_LEMMA = 2
i_POSTAG = 3
i_HEAD_ID = 4
i_DEPREL = 5

with open(collapsed_modified) as src:
    with open(training_regular_words_count_src, "w") as target:
        for line in src:
            line = line.split("\t")
            if len(line) <= 1:
                continue
            target.write( line[i_FORM] + " " + line[i_LEMMA] + "\n" )
            
```

```python
import subprocess, shlex, os

path_to_GloVe = "../Tools/GloVe-master/build/"
cmd = path_to_GloVe + "vocab_count"
cmd += " -min-count 1"
cmd += " -verbose 2"
with open(training_regular_words_count_src) as src:
    with open(training_regular_words_count, "w") as target:
        process = subprocess.Popen(shlex.split(cmd), stdin = src, stdout = target)
        (stdoutdata, stderrdata) = process.communicate()
```

Here we predefine the order of the embedded words as well as which words are contained within the embeddings in **word_to_index**

```python
replace_below_word_count = 5

number_of_placeholder_words = 15
placeholder_words = [ "<unknown>" ] + [ "<placeholder"+str(i)+">" for i in range(number_of_placeholder_words) ]

word_to_index = {}
with open(training_regular_words_count) as countfile:
    linenumber = 0
    for line in countfile:
        line = line.split()
        if int(line[1]) < replace_below_word_count:
            break
        word_to_index[line[0]] = linenumber
        linenumber += 1
```

- separator symbols, argument labels, and special relations are explicitly used for the output
- since a feed forward neural network learns the output index, it should not be a problem when the **indices of input and output are not equal**. Therefore we build different output word indices which include the special words. We do not need those special words for the input, thus we do not need actual GloVe embeddings for them
- add special word *empty* : 0

```python
# special words for the output
separators = ["<empty>", "<eos>",":(", "::", ",", ");"]
argument_labels = ['comp', 'condition', 'conj', 'dobj', 'iobj', 'mod', 'obj', 'outcome', 'poss', 'prop_of', 'sameAs_arg',
                   'source', 'subj', 'prep_', 'conj_']
special_relations = [ 'SameAs' ]


additional_words = separators + argument_labels + special_relations + placeholder_words
```

```python
word_to_index_regular_words = { }
embedding_matrix = None

with open(vocabulary_embeddings_regular_words) as embeddings_txt_file:
    
    # first line should contain dimensions
    dimensions_list = None
    embeddings_dimensions = None
    number_of_embeddings = None
    embedding_matrix = None
    
    linenumber = 0
    
    for line in embeddings_txt_file:
        
        # first line should contain dimensions
        if dimensions_list == None:
            dimensions_list = line.strip().split()
            embeddings_dimensions = int(dimensions_list[1])
            number_of_embeddings = len(word_to_index) + len(additional_words)
            embedding_matrix = numpy.zeros((number_of_embeddings, embeddings_dimensions), dtype = numpy.float32)
            continue
        
        line = line.strip().split()
        word = line[0]
        predefined_index = word_to_index.get(word, None) # we have to use none here instead of False due to index 0
        if predefined_index != None:
            word_to_index_regular_words[word]  = linenumber + len(additional_words)
            embedding_vector = line[1:]
            embedding_matrix[ word_to_index_regular_words[word] ] = map(float, embedding_vector)
            linenumber += 1
```

```python
mean = numpy.mean(embedding_matrix, axis=0, dtype=numpy.float32)
stddev = numpy.std(embedding_matrix, axis=0, dtype=numpy.float32)*0.01

for i in range(len(additional_words)):
    word_to_index_regular_words[  additional_words[i] ] = i

embedding_matrix[1,-1] = 1
for i in range(2,len(additional_words)):
    # randomly initialize the embeddings for the placeholder words for the input
    embedding_matrix[i] = numpy.random.normal(loc = mean, scale = stddev, size = (embeddings_dimensions) ) 
```

```python
with open(formatted_embeddings_regular_words, "w") as picklefile:
    pickle.dump((embedding_matrix, word_to_index_regular_words, number_of_placeholder_words),picklefile)
```

### POS Tags and Dependency Labels

```python
word_to_index_postags_deplabels = { "<empty>" : 0 }
embedding_matrix = None

with open(vocabulary_embeddings_postags_deplabels) as embeddings_txt_file:
    
    # first line should contain dimensions
    dimensions_list = embeddings_txt_file.readline().strip().split()
    embeddings_dimensions = int(dimensions_list[1])
    number_of_embeddings = int(dimensions_list[0]) + 1
    embedding_matrix = numpy.empty((number_of_embeddings, embeddings_dimensions))
    
    line_number = 0
    for line in embeddings_txt_file:
        line = line.strip().split()
        word = line[0]
        word_to_index_postags_deplabels[word] = line_number + 1
        
        embedding_vector = line[1:]
        embedding_matrix[line_number] = map(float, embedding_vector)
        line_number += 1
```

```python
with open(formatted_embeddings_postags_deplabels, "w") as picklefile:
    pickle.dump((embedding_matrix, word_to_index_postags_deplabels),picklefile)
```

# Format Training Data

- Transform training data to embedding indices
- We need to replace unknown words with placeholders, therefore we need to process each sample input in combination with its output

```python
X = []
Y = []

# conll-X table row elements
i_ID = 0
i_FORM = 1
i_LEMMA = 2
i_POSTAG = 3
i_HEAD_ID = 4
i_DEPREL = 5

issuecounter = 0

with open(collapsed_modified) as conll_file:
    with open(propsDE_extractions_modified) as extractions_file:
        sample = []
        sample_txt = [] # we need this for error recovery
        replacement_dict = {}
        for line in conll_file:
            line = line.strip().split()
            node = []
            if len(line) > 1:
            
                node.append( int(line[i_ID]) )
                if line[i_HEAD_ID] == '_':
                    node.append( 0 )
                else:
                    node.append( int(line[i_HEAD_ID]) )
                
                # we have to use none here instead of False due to index 0
                f = lambda w: word_to_index_regular_words.get(replacement_dict.get(w, w), None)
                
                wf = f(line[i_FORM])
                if wf == None:
                    if len(replacement_dict) < len(placeholder_words):
                        replacement_dict[ line[i_FORM] ] = placeholder_words[ len(replacement_dict) ]
                    else:
                        replacement_dict[ line[i_FORM] ] = placeholder_words[0]
                    wf = word_to_index_regular_words[ replacement_dict[ line[i_FORM] ] ]
                
                wl = f(line[i_LEMMA])
                if wl == None:
                    if len(replacement_dict) < len(placeholder_words):
                        replacement_dict[ line[i_LEMMA] ] = placeholder_words[ len(replacement_dict) ]
                    else:
                        replacement_dict[ line[i_LEMMA] ] = placeholder_words[0]
                    wl = word_to_index_regular_words[ replacement_dict[ line[i_LEMMA] ] ]
                
                
                node.append(wf)
                node.append(wl)
                sample_txt.append(line[i_FORM])
                sample_txt.append(line[i_LEMMA])
                
                node.append( word_to_index_postags_deplabels[line[i_POSTAG] ] )
                node.append( word_to_index_postags_deplabels[line[i_DEPREL] ] )
                sample.append(node)
            else:
                X.append(sample)
                sample = []
                
                line = extractions_file.readline().strip('\n').split()
                
                f = lambda w: word_to_index_regular_words.get(replacement_dict.get(w, w), None)
                
                # Find Issues
                restart = True
                while(restart):
                    restart = False
                    for i_w in range(len(line)):
                        w = line[i_w]
                        if f(w) == None:
                            issuecounter += 1
                            #print "ERROR word " + w + " line " + str(len(Y))
                        
                            # could be a concatenation of multiple words from the input
                            replace = []
                            for _ in range(2):
                                for word in sample_txt:
                                    if w.startswith(word):
                                        replace.append(word)
                                        w = w[len(word):]
                            #print "concat: " + ", ".join(replace)

                            # or it could be uppercase/lowercase variation from input
                            if len(replace) == 0:
                                for word in sample_txt:
                                    if w.lower() == word or word.lower() == w:
                                        replace.append(word)
                                        #print "replace: " + ", ".join(replace)
                                        break
                                        
                            line = line[:i_w] + replace + line[i_w+1:]
                            restart = True
                            break
                            
                Y.append( [ f(word) for word in line ] )            
                sample_txt = []
                replacement_dict = {}
                
print "Issues: " + str(issuecounter)
```

```python
with open(formatted_trainingdata, "w") as picklefile:
    pickle.dump((X, Y),picklefile)
```
